```python
import pandas as pd
```

```python
fb_df = pd.read_csv("dataset-fb-valence-arousal-anon.csv", header = 0)
```

```python
fb_df["Avg_Valence"] = fb_df[["Valence1", "Valence2"]].mean(axis = 1)
```

```python
fb_df["Avg_Arousal"] = fb_df[["Arousal1", "Arousal2"]].mean(axis = 1)
```

```python
fb_df.drop(columns = ["Valence1", "Valence2", "Arousal1", "Arousal2"], inplace = True)
```

```python
fb_df["Valence_Scaled"] = pd.cut(fb_df.Avg_Valence, bins=[0, 4.9, 5.9, 9.5] , labels=[-1.0, 0.0, 1.0], right=True)
```

```python
fb_df["Arousal_Scaled"] = pd.cut(fb_df.Avg_Arousal, bins = [0, 3.1, 4.6, 6.9, 9.5], labels=[-2.0, -1.0, 0.0, 1.0], right=True)
```
