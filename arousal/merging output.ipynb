```python
import pandas as pd
import numpy as np
from test2 import line_prepender
from test2 import clean_text
```

```python
out_original = pd.read_csv("output/outfinal.txt", header = 0)
out_descriptive = pd.read_csv("output/outfinal_descriptive_zeroshot.txt", header = 0)
```

```python
out_original.columns = ["Valence_original", "Arousal_original", "Phrase_text"]
```

```python
out_descriptive.columns = ["Valence_descriptive", "Arousal_descriptive", "Phrase_text"]
```

```python
merged = out_original.merge(out_descriptive.drop_duplicates(subset = ["Phrase_text"]), how = "left")
```

```python
merged["match"] = np.where(merged['Arousal_original'] == merged['Arousal_descriptive'], 1, 0)
```

```python
merged.drop(columns = ["Valence_descriptive"], inplace = True)
```
